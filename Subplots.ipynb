```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

plt.close('all')

# Just a figure and one subplot
f, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
plt.show()
```

```python
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].set_title('Sharing X axis')
axarr[0].plot(x, y)
axarr[1].scatter(x, y)

plt.show()
```

```python
# import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import svm
from sklearn.metrics import confusion_matrix
```

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2] # we only take the first two features. We could
 # avoid this ugly slicing by using a two-dim dataset
y = iris.target
```

```python
# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=1,gamma=1).fit(X, y)
```

```python
# np.meshgrid : https://stackoverflow.com/questions/36013063/what-is-purpose-of-meshgrid-in-python
# pandas indexing : http://pythonhow.com/accessing-dataframe-columns-rows-and-cells/
```

```python
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # x_min = 3.33 | x_max = 8.90 from First Col
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # y_min = 1.0  | y_max = 5.40 from Second Col
h = (x_max / x_min)/100 # 0.026969
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
```

```python
xx
```

```python
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()]) # c_ is used for concatenation | ravel() makes flat list
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)




plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
# plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```
